```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Function to modify HSV values and display images
def change_and_show_hsv_from_path(image_path, hue_shift=0, saturation_factor=1, value_factor=1):
    # Check if the image path exists
    if not os.path.exists(image_path):
        print(f"Error: The image at {image_path} does not exist.")
        return

    # Load the image
    image = cv2.imread(image_path)

    # Check if the image was loaded successfully
    if image is None:
        print(f"Error: Failed to load image at {image_path}")
        return

    # Convert the image from BGR to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Convert the RGB image to HSV
    image_hsv = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2HSV)

    # Modify the HSV values
    image_hsv[..., 0] = (image_hsv[..., 0] + hue_shift) % 180  # Hue (0-179)
    image_hsv[..., 1] = np.clip(image_hsv[..., 1] * saturation_factor, 0, 255)  # Saturation (0-255)
    image_hsv[..., 2] = np.clip(image_hsv[..., 2] * value_factor, 0, 255)  # Value (0-255)

    # Convert back to RGB
    modified_image_rgb = cv2.cvtColor(image_hsv, cv2.COLOR_HSV2RGB)

    # Display the original and modified image
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    # Original image
    axes[0].imshow(image_rgb)
    axes[0].set_title('Original Image')
    axes[0].axis('off')

    # Modified image
    axes[1].imshow(modified_image_rgb)
    axes[1].set_title(f'Modified Image (Hue: {hue_shift}, Sat: {saturation_factor}, Val: {value_factor})')
    axes[1].axis('off')

    plt.show()

# Example usage
image_path = '/content/drive/MyDrive/under_water_2/resize_500x500/858_img__500x500.png'  # Change this to the correct path

# Call the function with the image path
hue_shift = 10  # Change hue by 30
saturation_factor = 1.7  # Increase saturation by 20%
value_factor = 1.0  # Increase value by 10%

change_and_show_hsv_from_path(image_path, hue_shift, saturation_factor, value_factor)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Function to modify HSV values, split RGB channels and display images
def split_and_show_rgb_from_path(image_path):
    # Check if the image path exists
    if not os.path.exists(image_path):
        print(f"Error: The image at {image_path} does not exist.")
        return

    # Load the image
    image = cv2.imread(image_path)

    # Check if the image was loaded successfully
    if image is None:
        print(f"Error: Failed to load image at {image_path}")
        return

    # Split the image into its Red, Green, and Blue channels
    (B, G, R) = cv2.split(image)

    # Create images for each channel
    # Set the other channels to 0 to display the individual color channels
    red_channel = cv2.merge([np.zeros_like(R), np.zeros_like(R), R])
    green_channel = cv2.merge([np.zeros_like(G), G, np.zeros_like(G)])
    blue_channel = cv2.merge([B, np.zeros_like(B), np.zeros_like(B)])

    # Convert images from BGR to RGB for displaying
    red_channel_rgb = cv2.cvtColor(red_channel, cv2.COLOR_BGR2RGB)
    green_channel_rgb = cv2.cvtColor(green_channel, cv2.COLOR_BGR2RGB)
    blue_channel_rgb = cv2.cvtColor(blue_channel, cv2.COLOR_BGR2RGB)

    # Display the individual RGB channels
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))

    # Red channel image
    axes[0].imshow(red_channel_rgb)
    axes[0].set_title('Red Channel')
    axes[0].axis('off')

    # Green channel image
    axes[1].imshow(green_channel_rgb)
    axes[1].set_title('Green Channel')
    axes[1].axis('off')

    # Blue channel image
    axes[2].imshow(blue_channel_rgb)
    axes[2].set_title('Blue Channel')
    axes[2].axis('off')

    plt.show()

    # Optionally, save the channel images
    cv2.imwrite("/content/red_channel.png", red_channel)
    cv2.imwrite("/content/green_channel.png", green_channel)
    cv2.imwrite("/content/blue_channel.png", blue_channel)
    print("Channel images saved as red_channel.png, green_channel.png, blue_channel.png")

# Example usage
image_path = '/content/drive/MyDrive/under_water_2/enhance_images/Sharpening/858_img__500x500_Sharpening.png'  # Change this to the correct path

# Call the function with the image path
split_and_show_rgb_from_path(image_path)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Function to enhance and denoise individual RGB components
def enhance_and_denoise_rgb_component(image_path, component='red'):
    # Check if the image path exists
    if not os.path.exists(image_path):
        print(f"Error: The image at {image_path} does not exist.")
        return

    # Load the image
    image = cv2.imread(image_path)

    # Check if the image was loaded successfully
    if image is None:
        print(f"Error: Failed to load image at {image_path}")
        return

    # Split the image into its Red, Green, and Blue channels
    (B, G, R) = cv2.split(image)

    # Select the component to process
    if component == 'red':
        channel = R
    elif component == 'green':
        channel = G
    elif component == 'blue':
        channel = B
    else:
        print("Error: Invalid component choice. Choose 'red', 'green', or 'blue'.")
        return

    # Enhance the selected component (sharpening filter example)
    kernel_sharpen = np.array([[0, -1, 0], [-1, 5,-1], [0, -1, 0]])  # Simple sharpening kernel
    enhanced_channel = cv2.filter2D(channel, -1, kernel_sharpen)

    # Denoise the selected component (using Gaussian blur or bilateral filter)
    denoised_channel = cv2.GaussianBlur(enhanced_channel, (5, 5), 0)  # Example denoising with Gaussian blur

    # Create an image with the enhanced and denoised channel
    if component == 'red':
        enhanced_image = cv2.merge([np.zeros_like(R), np.zeros_like(R), enhanced_channel])
        denoised_image = cv2.merge([np.zeros_like(R), np.zeros_like(R), denoised_channel])
    elif component == 'green':
        enhanced_image = cv2.merge([np.zeros_like(G), enhanced_channel, np.zeros_like(G)])
        denoised_image = cv2.merge([np.zeros_like(G), denoised_channel, np.zeros_like(G)])
    elif component == 'blue':
        enhanced_image = cv2.merge([enhanced_channel, np.zeros_like(B), np.zeros_like(B)])
        denoised_image = cv2.merge([denoised_channel, np.zeros_like(B), np.zeros_like(B)])

    # Convert to RGB for displaying
    enhanced_image_rgb = cv2.cvtColor(enhanced_image, cv2.COLOR_BGR2RGB)
    denoised_image_rgb = cv2.cvtColor(denoised_image, cv2.COLOR_BGR2RGB)

    # Display the enhanced and denoised image
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    # Enhanced image
    axes[0].imshow(enhanced_image_rgb)
    axes[0].set_title(f'Enhanced {component.capitalize()} Component')
    axes[0].axis('off')

    # Denoised image
    axes[1].imshow(denoised_image_rgb)
    axes[1].set_title(f'Denoised {component.capitalize()} Component')
    axes[1].axis('off')

    plt.show()

    # Optionally, save the enhanced and denoised component images
    cv2.imwrite(f"/content/enhanced_{component}_component.png", enhanced_image)
    cv2.imwrite(f"/content/denoised_{component}_component.png", denoised_image)
    print(f"Enhanced and denoised {component} component images saved.")

# Example usage
image_path = '/content/drive/MyDrive/under_water_2/enhance_images/Sharpening/858_img__500x500_Sharpening.png'  # Change this to the correct path

# Call the function for 'red', 'green', or 'blue' component
component = 'red'  # Change this to 'green' or 'blue' for other components
enhance_and_denoise_rgb_component(image_path, component)
```
